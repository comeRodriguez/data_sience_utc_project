# ACM on variable 

```python
acm = pd.read_csv("../data/acm_data.csv").drop("Unnamed: 0", axis=1)
acm.columns = ["index", "X1", "X2", "X3", "X4", "X5", "X6", "X7", "X8", "X9", "X10", "color"]
acm
```

```python
i="X1"
j="X4"
plt.figure(figsize=(15, 15))
ax = sns.scatterplot(data=acm, x=i, y=j, style="color", hue="color", s=500)
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

label_point(acm[i], acm[j], acm["index"], plt.gca())  
plt.show()
```
