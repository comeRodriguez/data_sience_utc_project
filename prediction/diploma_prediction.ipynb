```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, CategoricalNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
clean_data = pd.read_csv("../data/clean_data.csv").drop("Unnamed: 0", axis=1)
clean_data
```

```python
clusters = pd.read_csv("../data/cluster_data.csv").drop("Unnamed: 0", axis=1)
clusters
```

```python
df = pd.merge(clean_data, clusters[["cluster", "payed"]], left_index=True, right_index=True, how="inner")
df
```

```python
sns.heatmap(clean_data.corr())
```

# KNN to determine salary of an individual

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    df.drop(["harmonized_salary"], axis=1),
    df[["harmonized_salary"]]
)
```

```python
X_train
```

```python
X_test["harmonized_salary"].mean()
```

```python
Y_train
```

```python
X_train_2 = pd.get_dummies(X_train[["cluster", "payed"]])
X_test_2 = pd.get_dummies(X_test[["cluster", "payed"]])
knn = KNeighborsRegressor()
knn.fit(X_train_2, Y_train)
```

```python
predictions = knn.predict(X_test_2)
```

```python
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

```python
r = Ridge()
r.fit(X_train_2, Y_train)
predictions = r.predict(X_test_2)
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

```python
l = Lasso()
l.fit(X_train_2, Y_train)
predictions = l.predict(X_test_2)
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

```python
e = ElasticNet()
e.fit(X_train_2, Y_train)
predictions = e.predict(X_test_2)
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

```python
lir = LinearRegression()
lir.fit(X_train_2, Y_train)
predictions = lir.predict(X_test_2)
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

```python
rfr = RandomForestRegressor()
rfr.fit(X_train_2, Y_train)
predictions = rfr.predict(X_test_2)
mse = mean_squared_error(predictions, Y_test)
np.sqrt(mse)
```

# Try to predict the category of salary

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    df.drop("payed", axis=1),
    df["payed"]
)
```

```python
X_train
```

```python
X_train_2 = pd.get_dummies(X_train[["cluster", "highest_level_of_education_completed", "gender", "years_of_experience_in_field", "industry"]])
X_test_2 = pd.get_dummies(X_test[["cluster", "highest_level_of_education_completed", "gender", "years_of_experience_in_field", "industry"]])
```

```python
X_train_2
```

```python
naives_bayes = CategoricalNB()
naives_bayes.fit(X_train_2, Y_train)
```

```python
predictions = naives_bayes.predict(X_test_2)
print(classification_report(Y_test, predictions, zero_division=True))
cm = confusion_matrix(Y_test, predictions, labels=naives_bayes.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=naives_bayes.classes_)
disp.plot()
plt.grid(False)
plt.show()
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train_2, Y_train)
predictions = tree.predict(X_test_2)
print(classification_report(Y_test, predictions, zero_division=True))
cm = confusion_matrix(Y_test, predictions, labels=tree.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=tree.classes_)
disp.plot()
plt.grid(False)
plt.show()
```

```python
rf = RandomForestClassifier()
rf.fit(X_train_2, Y_train)
predictions = rf.predict(X_test_2)
print(classification_report(Y_test, predictions, zero_division=True))
cm = confusion_matrix(Y_test, predictions, labels=rf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=rf.classes_)
disp.plot()
plt.grid(False)
plt.show()
```

```python
ada = AdaBoostClassifier()
ada.fit(X_train_2, Y_train)
predictions = ada.predict(X_test_2)
print(classification_report(Y_test, predictions, zero_division=True))
cm = confusion_matrix(Y_test, predictions, labels=ada.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=ada.classes_)
disp.plot()
plt.grid(False)
plt.show()
```
