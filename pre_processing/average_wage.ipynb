```python
df = df[df["currency"]!="Other"]
```

```python
exchange_rate = {
    "USD": 1,
    "GBP": 1.26,
    "CAD": 0.79,
    "EUR": 1.06,
    "AUD/NZD": 0.71,
    "CHF": 1.03,
    "ZAR": 0.065,
    "SEK": 0.1,
    "HKD": 0.13,
    "JPY": 0.0077
    
}
```

```python
def harmonize_salary(x):
    return x["annual_salary"] * exchange_rate[x["currency"]]
```

```python
df["harmonized_salary"] = df.apply(lambda x: harmonize_salary(x), axis=1)
df
```

```python
df = df[(df["harmonized_salary"] > df["harmonized_salary"].quantile(0.005)) & (df["harmonized_salary"] < df["harmonized_salary"].quantile(0.995))]
df
```

```python
df.to_csv("../data/survey_without_extreme_values.csv")
```
